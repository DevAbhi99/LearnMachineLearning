```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_csv("week2.csv")

x1=data.loc[:,'x1']
x2=data.loc[:,'x2']
y=data.loc[:,'y']

x=np.column_stack([x1,x2])

#Splitting feature for training

x_train=x[:799]

x_test=x[799:]

#splitting label for testing

y_train=y[:799]
y_test=y[799:]

#Creating a model

model=linear_model.LogisticRegression()


#training a model
model.fit(x_train,y_train)


#Testing the model
y_pred=model.predict(x_test)

#Finding mean square error

mse=mean_squared_error(y_test, y_pred)

print(f"mean squared error is {mse}")

#Finding coefficients and intercepts

```

```python
plt.scatter(y_test, y_pred, color='blue', label='Predicted vs Actual')  # Scatter plot

plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Perfect Prediction Line')

# Adding labels and title
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Actual vs Predicted Labels (Logistic Regression)')
plt.legend()

# Show the plot
plt.show()
```
